```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("telecom_customer_churn.csv")
```

```python
df.head()
```

```python
df.info()
```

#replacing blank spaces of "totalcharges" column values to 0 and converting the datatype to float

```python
df["TotalCharges"]=df["TotalCharges"].replace(" ","0")
df["TotalCharges"]=df["TotalCharges"].astype(float)
```

```python
df.info()
```

```python
df.isnull().sum().sum()
```

```python
df.duplicated().sum()
```

```python
df["customerID"].duplicated().sum()
```

```python
df.describe()
```

#converting seniorcitizen column data from boolen 1/0 to yess/no

```python
def conv(value):
    if value == 1:
        return "yes"
    else:
        return "No"
df['SeniorCitizen']=df['SeniorCitizen'].apply(conv)
```

```python
df.head()
```

```python
plt.figure(figsize=(4,4))
ax=sns.countplot(x='Churn',data=df)
plt.title("churn out count")
ax.bar_label(ax.containers[0])
sns.set_palette("muted")
plt.show()
```

#from above graph it is concluded that in 7043 customers 1869 customers are churned out and 5174 are still using the service 

```python
gb=df.groupby('Churn').agg({'Churn':"count"})
plt.figure(figsize=(3,3))
plt.title("churn result in percentage")
plt.pie(gb["Churn"],labels=gb.index,autopct="%1.2f%%")
```

#74% of customers are still using the service and 27% of customers are already churned out

```python
plt.figure(figsize=(4,4))
ax=sns.countplot(x="gender",data=df,hue="Churn")
ax.bar_label(ax.containers[0])
plt.title("customers churn by gender")
plt.show()
```

#by above visual we can conclude that the gender column does not have that much impact on churn rate

```python
plt.figure(figsize=(4,4))
ax=sns.countplot(x='SeniorCitizen',data=df)
plt.title("senior citizen count")
ax.bar_label(ax.containers[0])
plt.show()
```

#in total 7043 customers 1142 customers are seniorcitizens

```python
plt.figure(figsize=(3,4))
ax = sns.countplot(x="SeniorCitizen", data=df, hue="Churn")
ax.bar_label(ax.containers[0])
ax.bar_label(ax.containers[1])
plt.title("Customers Churn by SeniorCitizen")
plt.show()
```

#from above visual we can conclude that most of the customers from seniorcitizens are churned out compared to other customers

```python
plt.figure(figsize=(9,4))
sns.histplot(x='tenure',data=df,bins=72,hue='Churn')
plt.show()
```

#Churn is highest among customers with shorter tenure.

#Customers who have been with the company for a long time tend to stay.



```python
ax=sns.countplot(x='Contract',data=df,hue='Churn')
ax.bar_label(ax.containers[0])
ax.bar_label(ax.containers[1])
plt.title("Customers Churn by Contract")
plt.show()
```

#churn out rate of customers is high who has taken month-to-month contract as compared to one year contract and two year contract

```python
features = ['PhoneService', 'MultipleLines', 'InternetService',
            'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
            'TechSupport', 'StreamingTV', 'StreamingMovies']

plt.figure(figsize=(18, 12))
sns.set_palette("muted")  #Soft & aesthetic color palette

for i, col in enumerate(features, 1):
    ax = plt.subplot(3, 3, i)
    sns.countplot(x=col, hue='Churn', data=df, ax=ax)
    ax.set_title(f'Churn by {col}', fontsize=10)
    ax.set_xlabel('')
    ax.set_ylabel('Count')
    ax.tick_params(axis='x', rotation=30)

    # Add count labels
    for p in ax.patches:
        if p.get_height() > 0:
            ax.text(p.get_x() + p.get_width()/2, p.get_height() + 2,
                    int(p.get_height()), ha='center', fontsize=8)

plt.tight_layout()
plt.show()
```

#Customers without internet-based services (like OnlineSecurity, Backup, etc.) generally show lower churn.

#For features like OnlineSecurity, TechSupport, and DeviceProtection, those who don’t have the service are more likely to churn.

#Fiber Optic users show higher churn compared to DSL and No internet.

#Phone and streaming services show a visible churn difference but are less impactful compared to internet-based features.
